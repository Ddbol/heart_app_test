```python
## defining independent and dependent variables
x=dfheart.drop(["DEATH_EVENT"],axis=1)
y=dfheart["DEATH_EVENT"]
```

```python

scaler = StandardScaler()

# Fit and transform the data
scale_x = scaler.fit_transform(x)


# Save the scaler to a file using pickle
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

print("Scaler has been re-saved successfully!")
```

```python
print(type(scaler))
```

```python
## train-test split 
x_train, x_test, y_train,y_test = train_test_split(scale_x,y,test_size=0.4,random_state=23)
```

```python
# Instantiating the SVM algorithm 
#Create a SVM Classifier
svm_linear = svm.SVC(kernel='linear')
svm_sigmoid = svm.SVC(kernel='sigmoid') 
svm_rbf = svm.SVC(kernel='rbf') 

# Fitting the model 
svm_linear.fit(x_train, y_train)
svm_sigmoid.fit(x_train, y_train)
svm_rbf.fit(x_train, y_train)

# Predicting the test variables
y_pred_linear = svm_linear.predict(x_test)
y_pred_sigmoid = svm_sigmoid.predict(x_test)
y_pred_rbf = svm_rbf.predict(x_test)

# Getting the score 
print("Linear Kernel Accuracy:",metrics.accuracy_score(y_test, y_pred_linear))
print("Sigmoid kernel Accuracy:",metrics.accuracy_score(y_test, y_pred_sigmoid))
print("rbf kernelAccuracy:",metrics.accuracy_score(y_test, y_pred_rbf))
```

```python
# Save the svm model
with open('svm.pkl', 'wb') as f:
    pickle.dump(svm_linear, f)

print("model has been re-saved successfully!")
```

```python
## Neural Network
## initialise the NN
# pass series of layers in a fixed order
model = Sequential()

##set up the input/visible layers - 1D array with 12 elements
model.add(Input(shape=(12,)))
# could alternatively use this instead of 12 #x.shape[1]

## input layer - dense type of nn, aka fully connected layout. units is number of neurons. kernel initialiser setting up weight (uniform randomly assign weight from uniform distribution)
# keras documentation different kernel_initialisations e.g. specify min and max value
# these input neurons are self defined (can be any number)
"""you can tell the first one is the input layer because it is 'model.add(Input...)' whereas 1st hidden layer is 'model.Dense(...)', as are all the subsequent hidden layers
 
so all hidden layers are model.add(Dense...)
 
model = nn.Sequential(
    nn.Linear(12, 50),  # 12 input features, 50 output features
    nn.Linear(50, 10)   # subsequent layer, takes previous layer's output
)
 
It's like 12 = number of features in X.
output of first layer is passed as input to second layer (both must be 50)
Outputs layer size 10 at the end.
 
PyTorch implementation of same idea if this clears anything up for people
 
if curious
 """


# level complexity determines number of hidden layers

model.add(Dense(units= 50, kernel_initializer= 'uniform', activation = 'relu'))
## hidden layer 1
model.add(Dense(units= 20, kernel_initializer= 'uniform', activation = 'relu'))
##regularization 1
model.add(Dropout(0.4))
## hidden layer 2
model.add(Dense(units= 4, kernel_initializer= 'uniform', activation = 'relu'))
##regularization 2
model.add(Dropout(0.2))
## 0,1,2,3,4 ---- softmax
##output layer
model.add(Dense(units= 1, kernel_initializer= 'uniform', activation = 'sigmoid'))

## compiling the ANN
## there is also rmsprop optimiser. adam update model weight at point of training, compares y red adn y actual, predicting to classes so use binar_crossentroupy, if multiclass categorical_crossentropy
# can use precision or recall for metrics etc
model.compile(optimizer = 'adam', loss ='binary_crossentropy', metrics = ['accuracy'])

# train, test and validation data if have lots of data. don't have lots of data use 25%
## model
history = model.fit(x_train, y_train, batch_size = 25, epochs = 80, validation_split=0.25)

# accuracy val_accuracy validatoin split or could use test data set
# does take into account validation and apply changes????

# if get same with just input and output layer - less complex don't need hidden layers

# potentially overfitting the training as the validation so far apart - simplify network (reduce hidden layers) reduce epochs or do early stopping (stop when network not learning anymore)
```
